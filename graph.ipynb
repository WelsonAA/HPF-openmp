```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Organized Data
benchmark_data = {
    "360p": {
        1: 266.4,
        2: 171.6,
        3: 137.4,
        4: 117.1,
        5: 119.5,
        6: 119.8,
        7: 122.4,
        8: 119.6,
    },
    "480p": {
        1: 428.8,
        2: 254.0,
        3: 210.6,
        4: 168.9,
        5: 162.0,
        6: 165.2,
        7: 161.4,
        8: 161.5,
    },
    "720p": {
        1: 893.1,
        2: 497.7,
        3: 365.2,
        4: 290.9,
        5: 292.7,
        6: 292.1,
        7: 289.3,
        8: 298.1,
    },
    "1080p": {
        1: 1937,
        2: 1029,
        3: 722.5,
        4: 571.6,
        5: 568.4,
        6: 578.6,
        7: 568.2,
        8: 574.6,
    },
    "2K": {
        1: 6687,
        2: 3481,
        3: 2389,
        4: 1867,
        5: 1866,
        6: 1872,
        7: 1898,
        8: 1859,
    },
    "4K": {
        1: 26624,
        2: 13800,
        3: 9301 ,
        4: 7268 ,
        5: 7187 ,
        6: 7149 ,
        7: 7148 ,
        8: 7179 ,
    }
}

size_order = ["360p", "480p", "720p", "1080p", "2K", "4K"][::-1]

# Flatten data for plotting
sizes = []
nodes = []
times = []

for size in size_order:
    for node, time in benchmark_data[size].items():
        sizes.append(size_order.index(size))
        nodes.append(node)
        times.append(time)

# Convert data to numpy arrays
sizes = np.array(sizes)
nodes = np.array(nodes)
times = np.array(times)

# Create 3D plot
fig = plt.figure(figsize=(10, 10))

ax = fig.add_subplot(111, projection='3d')

# Create 3D bars
width = 0.9  # Width of the bars
depth = width  # Depth of the bars
ax.bar3d(sizes, nodes, np.zeros_like(times), width,
         depth, times, shade=True, color='skyblue')

# Set categorical axis labels
ax.set_xticks(range(len(size_order)))
ax.set_xticklabels(size_order)
ax.set_yticks(range(1, 9))
ax.set_yticklabels(range(1, 9))

# Remove gridlines
ax.grid(False)

# Rotate the plot
# ax.view_init(30, 135)
ax.view_init(elev=30, azim=-60)

# Label axes
ax.set_xlabel('Image Size')
ax.set_ylabel('Number of Nodes')
ax.set_zlabel('Time Taken (ms)')

# Add margin to plot
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)

# Show plot
plt.title("Benchmarking Results")
plt.show()

# Save plot
fig.savefig("benchmark.jpg", dpi=150)
```
